# GOOD LOAN METRICES

```python
good_loans = df[df['loan_status'].isin(["Fully Paid", "Current"])]
total_loan_applications = df['id'].count()
good_loan_applications = good_loans['id'].count()
good_loan_funded_amount = good_loans['loan_amount'].sum()
good_loan_recieved_amount = good_loans['total_payment'].sum()

good_loan_funded_amount_millions = good_loan_funded_amount/1000000
good_loan_recieved_amount_millions = good_loan_recieved_amount/1000000

good_loan_percentage = (good_loan_applications/total_loan_applications)*100

print("GOOD LOAN APPLICATIONS:", good_loan_applications)
print("GOOD LOAN FUNDED AMOUNT: ${:.2f}M".format(good_loan_funded_amount_millions))
print("GOOD LOAN RECEIVED AMOUNT: ${:.2f}M".format(good_loan_recieved_amount_millions))
print("GOOD LOAN PERCENTAGE: {:.2f}%".format(good_loan_percentage))
```

# BAD LOAN METRICES

```python
bad_loans = df[df['loan_status'].isin(["Charged Off"])]
total_loan_applications = df['id'].count()
bad_loan_applications = bad_loans['id'].count()
bad_loan_funded_amount = bad_loans['loan_amount'].sum()
bad_loan_recieved_amount = bad_loans['total_payment'].sum()

bad_loan_funded_amount_millions = bad_loan_funded_amount/1000000
bad_loan_recieved_amount_millions = bad_loan_recieved_amount/1000000

bad_loan_percentage = (bad_loan_applications/total_loan_applications)*100

print("BAD LOAN APPLICATIONS:", bad_loan_applications)
print("BAD LOAN FUNDED AMOUNT: ${:.2f}M".format(bad_loan_funded_amount_millions))
print("BAD LOAN RECEIVED AMOUNT: ${:.2f}M".format(bad_loan_recieved_amount_millions))
print("BAD LOAN PERCENTAGE: {:.2f}%".format(bad_loan_percentage))
```

# MONTHLY TRENDS BY ISSUE DATE FOR TOTAL FUNDED AMOUNT

```python
monthly_funded = (
    df.sort_values('issue_date')
    .assign(month_name = lambda x:x['issue_date'].dt.strftime('%B, %Y'))
    .groupby('month_name', sort = False)['loan_amount']
    .sum()
    .div(1000000)
    .reset_index(name='loan_amount_millions')
)

plt.figure(figsize=(12,6))
plt.fill_between(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], color = 'skyblue', alpha = 0.5)
plt.fill(monthly_funded['month_name'], monthly_funded['loan_amount_millions'], color = 'blue', linewidth = 2)

for i, row in monthly_funded.iterrows():
    plt.text(i, row['loan_amount_millions'] + 0.1, f"{row['loan_amount_millions']:.2f}",
             ha = 'center', va = 'bottom', fontsize = 9, rotation = 0, color = 'black')

plt.title('Total Funded Amount by Month', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Funded Amount (Millions)')
plt.xticks(ticks=range(len(monthly_funded)), labels = monthly_funded['month_name'], rotation = 90)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.tight_layout()
plt.show()
```

# MONTHLY TRENDS BY ISSUE DATE FOR TOTAL AMOUNT RECIEVED

```python
monthly_recieved = (
    df.sort_values('issue_date')
    .assign(month_name = lambda x:x['issue_date'].dt.strftime('%B, %Y'))
    .groupby('month_name', sort = False)['total_payment']
    .sum()
    .div(1000000)
    .reset_index(name='recieved_amount_millions')
)

plt.figure(figsize=(12,6))
plt.fill_between(monthly_recieved['month_name'], monthly_recieved['recieved_amount_millions'], color = 'green', alpha = 0.5)
plt.fill(monthly_recieved['month_name'], monthly_recieved['recieved_amount_millions'], color = 'yellow', linewidth = 2)

for i, row in monthly_recieved.iterrows():
    plt.text(i, row['recieved_amount_millions'] + 0.1, f"{row['recieved_amount_millions']:.2f}",
             ha = 'center', va = 'bottom', fontsize = 9, rotation = 0, color = 'black')

plt.title('Total Recieved Amount by Month', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Recieved Amount (Millions)')
plt.xticks(ticks=range(len(monthly_recieved)), labels = monthly_recieved['month_name'], rotation = 90)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.tight_layout()
plt.show()
```

# MONTHLY TRENDS BY ISSUE DATE FOR TOTAL LOAN APPLICATIONS

```python
monthly_applications = (
    df.sort_values('issue_date')
    .assign(month_name = lambda x:x['issue_date'].dt.strftime('%B, %Y'))
    .groupby('month_name', sort = False)['id']
    .count()
    .reset_index(name='loan_application_count')
)

plt.figure(figsize=(12,6))
plt.fill_between(monthly_applications['month_name'], monthly_applications['loan_application_count'], color = 'red', alpha = 0.5)
plt.fill(monthly_applications['month_name'], monthly_applications['loan_application_count'], color = 'orange', linewidth = 2)

for i, row in monthly_applications.iterrows():
    plt.text(i, row['loan_application_count'] + 0.1, f"{row['loan_application_count']}",
             ha = 'center', va = 'bottom', fontsize = 9, rotation = 0, color = 'black')

plt.title('Total Loan Applications by Month', fontsize = 14)
plt.xlabel('Month')
plt.ylabel('Number of Applications')
plt.xticks(ticks=range(len(monthly_applications)), labels = monthly_applications['month_name'], rotation = 90)
plt.grid(True, linestyle = '--', alpha = 0.6)
plt.tight_layout()
plt.show()
```

# REGIONAL ANALYSIS BY STATE FOR TOTAL FUNDED AMOUNT

```python
import matplotlib.pyplot as plt

# Group by state and sum loan_amount (in thousands)
state_funding = df.groupby('address_state')['loan_amount'].sum().sort_values(ascending=True)
state_funding_thousands = state_funding / 1000

plt.figure(figsize=(10, 8))

bars = plt.barh(state_funding_thousands.index, state_funding_thousands.values, color='lightcoral')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 10, bar.get_y() + bar.get_height() / 2, 
             f'{width:,.0f}K', va='center', fontsize=9)

plt.title('Total Funded Amount by State (in Thousands)')
plt.xlabel('Funded Amount (in Thousands)')
plt.ylabel('State')
plt.tight_layout()
plt.show()
```

# LONG TERM ANALYSIS BY TOTAL FUNDED AMOUNT

```python
import matplotlib.patches as patches
# Group by loan term and sum loan_amount (in millions)
term_funding_millions = df.groupby('term')['loan_amount'].sum() / 1_000_000

plt.figure(figsize=(5,5))
plt.pie(
    term_funding_millions,
    labels=term_funding_millions.index,
    autopct=lambda p: f"{p:.1f}%\n${p*sum(term_funding_millions)/100:.1f}M",
    startangle=90,
    wedgeprops={'width':0.4}
)

# Add white circle to create "donut chart"
centre_circle = patches.Circle((0,0), 0.70, color='white', fc='white')
plt.gca().add_artist(centre_circle)

plt.title("Total Funded Amount By Term (in million)")
plt.show()
```

# EMPLOYEE LENGTH BY TOTAL AMOUNT FUNDING

```python
import matplotlib.pyplot as plt

# Group by employee length and sum loan_amount (in thousands)
emp_funding = df.groupby('emp_length')['loan_amount'].sum() / 1000  

plt.figure(figsize=(10,6))

# Use emp_funding instead of emp_funding_thousands
bars = plt.barh(emp_funding.index, emp_funding, color='blue')

# Add value labels
for bar in bars:
    width = bar.get_width()
    plt.text(width + 5, bar.get_y() + bar.get_height() / 2, 
             f'{width:,.0f}K', va='center', fontsize=9)

plt.title('Total Funded Amount by Employee Length')
plt.xlabel('Funded Amount (in Thousands)')
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

# LOAN PURPOSE BY TOTAL FUNDED AMOUNT

```python
# Group by purpose and sum loan_amount (in thousands or millions)
purpose_funding = df.groupby('purpose')['loan_amount'].sum().sort_values(ascending=False)
purpose_funding_millions = purpose_funding / 1_000_000   # convert to millions

plt.figure(figsize=(12, 6))

bars = plt.bar(purpose_funding_millions.index, purpose_funding_millions.values, color='teal')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.5,
             f'{height:,.1f}M', ha='center', va='bottom', fontsize=9)

plt.title('Total Funded Amount by Loan Purpose (in Millions)')
plt.xlabel('Loan Purpose')
plt.ylabel('Funded Amount (in Millions)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

# HOME OWNERSHIP BY TOTAL FUNDED AMOUNT

```python
# Group by home ownership and sum loan_amount (convert to millions for readability)
home_funding = df.groupby('home_ownership')['loan_amount'].sum().sort_values(ascending=False)
home_funding_millions = home_funding / 1_000_000   # in millions

plt.figure(figsize=(8,5))

bars = plt.bar(home_funding_millions.index, home_funding_millions.values, color='royalblue')

# Add value labels
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 0.3,
             f'{height:,.1f}M', ha='center', va='bottom', fontsize=9)

plt.title('Total Funded Amount by Home Ownership (in Millions)')
plt.xlabel('Home Ownership')
plt.ylabel('Funded Amount (in Millions)')
plt.tight_layout()
plt.show()

import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Group by home ownership and sum loan_amount (millions)
home_funding = df.groupby('home_ownership')['loan_amount'].sum().sort_values(ascending=False)
home_funding_millions = home_funding / 1_000_000

plt.figure(figsize=(7,7))

# Donut chart with only percentage labels
wedges, texts, autotexts = plt.pie(
    home_funding_millions,
    labels=home_funding_millions.index,
    autopct='%1.1f%%',
    startangle=90,
    textprops={'fontsize':9},
    wedgeprops={'width':0.4}
)

# Add white circle for donut effect
centre_circle = patches.Circle((0,0),0.70,color='white')
plt.gca().add_artist(centre_circle)

# Add legend with exact amounts
plt.legend(
    [f"{idx}: ${val:,.1f}M" for idx, val in zip(home_funding_millions.index, home_funding_millions.values)],
    title="Funded Amount",
    loc="center left",
    bbox_to_anchor=(1, 0.5)
)

plt.title("Total Funded Amount by Home Ownership (in Millions)")
plt.tight_layout()
plt.show()
```
